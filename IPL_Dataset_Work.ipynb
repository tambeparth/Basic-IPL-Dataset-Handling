```python
# Extracting top 5 players and their corresponding counts
top_players = ipl['Player_of_Match'].value_counts().head(5)
player_names = top_players.index.tolist()
player_counts = top_players.values.tolist()

# Plotting the bar chart
plt.figure(figsize=(8, 5))
plt.bar(player_names, player_counts, color='g')
plt.xlabel('Player')
plt.ylabel('Number of Player of Match Awards')
plt.title('Top 5 Players with Most Player of Match Awards')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
batting_first=ipl[ipl['WonBy']!=0]
```

```python
batting_first.head()
```

```python
plt.figure(figsize=(5,7))
plt.hist(batting_first['WonBy'])
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
batting_first['WinningTeam'].value_counts()
```

```python
plt.figure(figsize=(6, 6))
batting_first_counts = batting_first['WinningTeam'].value_counts()
plt.bar(batting_first_counts.index[:3], batting_first_counts[:3], color=["blue", "yellow", "orange"])
plt.show()
```

```python
plt.figure(figsize=(6, 6))
batting_first_counts = batting_first['WinningTeam'].value_counts()[:3]
plt.pie(batting_first_counts, labels=batting_first_counts.index, colors=["blue", "yellow", "orange"], autopct='%1.1f%%', startangle=140)
plt.title('Top 3 Winning Teams When Batting First')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(8, 8))
team_wins = ipl['WinningTeam'].value_counts()
plt.pie(team_wins, labels=team_wins.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Wins by Team')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
